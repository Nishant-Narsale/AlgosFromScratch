```python
data = np.array(data)
print("Data Shape : ", data.shape)
m, n = data.shape

#shuffle data to divide it into two datasets
np.random.shuffle(data)

#validation data
data_val = data[0:1000].T   #First row will be the output, and that column will represent the pixel values due to Transpose
y_val = data_val[0]
X_val = data_val[1:]
print("X_val : ", X_val.shape)
print("y_val : ", y_val.shape)

#training data
data_train = data[1000:].T     #Each column represent each image due to Transpose
y_train = data_train[0]
X_train = data_train[1:]
print("X_train : ", X_train.shape)
print("y_train : ", y_train.shape)
```

```python
def initialize_learnable_params():
    w1 = np.random.uniform(low=-0.5, high=0.5, size=(10,784))  #between -0.5 and 0.5
    b1 = np.random.uniform(low=-0.5, high=0.5, size=(10, 1))
    w2 = np.random.uniform(low=-0.5, high=0.5, size=(10, 10))
    b2 = np.random.uniform(low=-0.5, high=0.5, size=(10, 1))

    return w1, b1, w2, b2

w1, b1, w2, b2 = initialize_learnable_params()
```

```python
def softmax(Z):
    A = np.exp(Z - np.max(Z)) / sum(np.exp(Z))
    return A

def relu(z):
    return np.maximum(0, z)

def forward(X_train, w1, b1, w2, b2):
    print("w1 :", w1.shape, "w2 :", w2.shape, "b1 :", b1.shape, "b2 :", b2.shape, "X_train :", X_train.shape)
    z1 = np.dot(w1, X_train) + b1
    a1 = relu(z1)
    z2 = np.dot(w2, a1) + b2
    print(z2.T.shape)
    a2 = softmax(z2.T)
    print(a2.shape)
    print(a2[0])

# w1, b1, w2, b2 = initialize_learnable_params()
forward(X_train, w1, b1, w2, b2)
```
